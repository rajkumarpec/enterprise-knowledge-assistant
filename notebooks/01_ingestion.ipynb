```python
#ingestion all pdfs

RAW_DOCS_DIR = Path("data/documents")

all_documents = []

for pdf_path in RAW_DOCS_DIR.glob("*.pdf"):
    print(f"\n--- Processing: {pdf_path.name} ---")
    docs = ingest_pdf(
        pdf_path,
        show_preview=False,   # keep clean for batch
        verbose=True
    )
    all_documents.extend(docs)

print(f"\nâœ… Total pages ingested from all PDFs: {len(all_documents)}")
```
