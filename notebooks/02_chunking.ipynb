```python
# import modules
from pathlib import Path

from src.ingestion.pdf_loader import ingest_pdf
from src.chunking.text_chunker import chunk_documents

```

```python
# Execute Step-1 (Ingestion)

RAW_DOCS_DIR = Path("data/documents")

documents = []

for pdf_path in RAW_DOCS_DIR.glob("*.pdf"):
    print(f"\nğŸ“„ Ingesting PDF: {pdf_path.name}")
    docs = ingest_pdf(
        pdf_path,
        show_preview=False,
        verbose=True
    )
    documents.extend(docs)

print(f"\nâœ… Total pages ingested from all PDFs: {len(documents)}")

```

```python
# Execute Step-2 (Chunking)

chunked_documents = chunk_documents(
    documents,
    chunk_size=1000,
    chunk_overlap=200,
    preview_chunks=3,
    verbose=True
)

```

```python
# ===== Final Chunk Validation =====

print("ğŸ” Sanity Check on First Chunk\n")

first_chunk = chunked_documents[0]

print("PDF Source :", first_chunk.metadata.get("source"))
print("Page       :", first_chunk.metadata.get("page"))
print("File Path  :", first_chunk.metadata.get("file_path"))

print("\nğŸ“ƒ Chunk Text Preview:\n")
print(first_chunk.page_content[:400])

```

```python
# Chunk Statistics (Optional but Useful)

from collections import Counter

pdf_counter = Counter(
    chunk.metadata["source"] for chunk in chunked_documents
)

print("ğŸ“Š Chunk distribution by PDF:\n")
for pdf, count in pdf_counter.items():
    print(f"{pdf}: {count} chunks")

print(f"\nğŸ“¦ Total chunks created: {len(chunked_documents)}")
```
